## Training and Evaluation

```python
x=df.drop(columns=['Price(USD)'])
y=df['Price(USD)']
```

```python
x_s = preprocessing.StandardScaler().fit(x).transform(x.astype(float))
x_s[0:5]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_s_train,x_s_test,y_train,y_test=train_test_split(x_s,y,train_size=0.8,random_state=4)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=4)
```

# Linear Regression

```python
reg=LinearRegression()
reg.fit(x_s_train,y_train)
reg.coef_
```

```python
from sklearn.metrics import r2_score
r2_score_Linear=r2_score(y_test,reg.predict(x_s_test))
print('R2_score: ',r2_score_Linear)
print('MSE: ',np.mean((y_test-reg.predict(x_s_test))**2))
print('MAE: ',np.mean(np.abs(y_test-reg.predict(x_s_test))))
```

```python
from sklearn.ensemble import RandomForestRegressor
freg=RandomForestRegressor()
freg.fit(x_train,y_train)
```

```python
r2_score_RF=r2_score(y_test,freg.predict(x_test))
print('R2_score: ',r2_score_Linear)
print('MSE: ',np.mean((y_test-freg.predict(x_test))**2))
print('MAE: ',np.mean(np.abs(y_test-freg.predict(x_test))))
```

```python
scores={r2_score_Linear:'Linear',r2_score_RF:'RF'}
The_Best=max(scores)
print('The best model for this problem is ',scores[The_Best],' with R2_score of ',The_Best)
```
